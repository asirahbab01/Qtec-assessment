```python
#Importing necessary libraries

from collections import Counter
import re
import os
import nltk
from nltk.corpus import stopwords
import numpy as np
# Download the stopwords from NLTK (if not already downloaded)
nltk.download('stopwords')

# Step 1: To read the chat logs from a specific folder
FOLDER_PATH = 'chat_logs'
STOPWORDS = set(stopwords.words('english'))
```

```python
# Step 2: Parse the chat logs

# Read uploaded file
for file_name in os.listdir(FOLDER_PATH):
    if file_name.endswith('.txt'):
        file_path = os.path.join(FOLDER_PATH, file_name) # Gets the first uploaded file
    with open(file_path, 'r') as file:
        lines = file.readlines()

user_messages = []
ai_messages = []

# Separate messages by speaker
for line in lines:
    line = line.strip()
    if line.startswith("User:"):
        user_messages.append(line.replace("User:", "").strip())
    elif line.startswith("AI:"):
        ai_messages.append(line.replace("AI:", "").strip())
```

```python
# Step 3: Message statistics
total_messages = len(user_messages) + len(ai_messages)
user_count = len(user_messages)
ai_count = len(ai_messages)

# Step 4: Display results
print("\n--- Chat Log Parsing Complete ---")
print(f"Total Messages: {total_messages}")
print(f"User Messages: {user_count}")
print(f"AI Messages: {ai_count}")
```

```python
# Step 5 Keyword Analysis

# Define English stopwords
stop_words = set(stopwords.words('english'))

# Combine all messages
all_text = " ".join(user_messages + ai_messages)

# Tokenize and clean text
tokens = re.findall(r'\b\w+\b', all_text.lower())  # Word tokens, lowercase
filtered_tokens = [word for word in tokens if word not in stop_words]

# Count word frequencies
word_freq = Counter(filtered_tokens)
top_keywords = [word for word, freq in word_freq.most_common(5)]
```

```python
# Step 6: Generate Summary

summary = f"""
--- Chat Summary ---
Total Exchanges: {len(user_messages) + len(ai_messages)}
User Messages: {len(user_messages)}
AI Messages: {len(ai_messages)}


Example Summary:
- The conversation had {len(user_messages) + len(ai_messages)} exchanges.
- The user asked mainly about {top_keywords[0]} and related topics.
- Most common keywords: {', '.join(top_keywords)}.
"""

# Display Summary
print(summary)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
# Bonus: TF-IDF Keyword Extraction

# Combine all messages into one string (treated as one document)
documents = [" ".join(user_messages + ai_messages)]

# Vectorize with TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

# Extract top 5 keywords by TF-IDF score
feature_array = np.array(vectorizer.get_feature_names_out())
tfidf_sorting = np.argsort(tfidf_matrix.toarray()).flatten()[::-1]
top_keywords_tfidf = feature_array[tfidf_sorting][:5]

# Show result
print("\n--- TF-IDF Keywords ---")
print("Top Keywords (TF-IDF):", ", ".join(top_keywords_tfidf))
```

```python
# Summarization of multiple chat logs

print("Processing all chat log (.txt) files in the folder:", FOLDER_PATH)

# Prepare a list to collect summaries
all_summaries = []

# Loop through each .txt file in the folder
for file_name in os.listdir(FOLDER_PATH):
    if file_name.endswith('.txt'):
        file_path = os.path.join(FOLDER_PATH, file_name)
        with open(file_path, 'r') as f:
            lines = f.readlines()

        # Parse messages
        user_msgs = []
        ai_msgs = []

        for line in lines:
            line = line.strip()
            if line.startswith("User:"):
                user_msgs.append(line.replace("User:", "").strip())
            elif line.startswith("AI:"):
                ai_msgs.append(line.replace("AI:", "").strip())

        # Combine and clean text
        full_text = " ".join(user_msgs + ai_msgs)
        tokens = re.findall(r'\b\w+\b', full_text.lower())
        clean_tokens = [word for word in tokens if word not in stop_words]
        word_counts = Counter(clean_tokens)
        common_keywords = [word for word, _ in word_counts.most_common(5)]

        # Summary per file
        summary = f"""
File: {file_name}
- Total Exchanges: {len(user_msgs) + len(ai_msgs)}
- User Messages: {len(user_msgs)}
- AI Messages: {len(ai_msgs)}
- Most Common Keywords: {', '.join(common_keywords)}
"""
        print(summary)
        all_summaries.append(summary)

# Optionally save all summaries to a single file
with open("multi_chat_summary.txt", "w") as f:
    for s in all_summaries:
        f.write(s + "\n")

print("Summaries saved to multi_chat_summary.txt")
```
